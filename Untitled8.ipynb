```python
def fit(X_train, y_train, X_val, y_val, learning_rate, max_epochs):
    '''
    Perform gradient descent on given Training Data for different learning rate
    and number of epochs

    Returns-
    1. Weights at the end of convergence
    2. Final RMSE on train set
    3. Epochs required to converge
    '''

    train_examples, number_of_features1 = X_train.shape
    val_example, number_of_features2 = X_val.shape
    weights = np.zeros((number_of_features1, 1))
    loss = 1e10
    epochs = 0
    train_rmse = 0
    train_loss, val_loss = [], []

    while True:
        predictions = X_train @ weights
        differences = np.subtract(predictions, y_train)
        differences_square = np.square(differences)
        loss = np.sum(differences_square)/(train_examples)
        train_loss.append(loss)
        train_rmse = math.sqrt(loss)

        predictions_val = X_val @ weights
        differences_val = np.subtract(predictions_val, y_train)
        differences_square_val = np.square(differences_val)
        loss_val = np.sum(differences_square_val)/(val_examples)
        val_loss.append(loss_val)
        val_rmse = math.sqrt(loss_val)

        # Terminating condition:
        if epochs >= max_epochs:
            break

        dj = (X_train.T @ differences)/(train_examples)
        weights = weights - learning_rate*dj
        epochs += 1

    plt.plot(np.arrange(len(train_loss)), train_loss, 'blue')
    plt.plot(np.arrange(len(val_loss)), val_loss, 'red')
    plt.legend(['Train RMSE', 'Val RMSE'])
    plt.xlabel("Epochs")
    plt.ylabel("RMSE")
    plt.show()

    return weights, epochs, train_rmse, val_rmse
```

```python
import math

def predict(X_test, y_test, weights):
    predicted = X_test @ weights
    difference = np.subtract(predicted, y_test)
    difference_square = np.square(difference)
    mse = np.mean(difference_square)
    rmse = math.sqrt(mse)
    return rmse
```
