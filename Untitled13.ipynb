```python
"""The np.concatenate() function expects a sequence of arrays 
(like a list or tuple) to concatenate along a specified axis. However, 
you're passing two separate arguments instead of a sequence. 

Also, the np.ones(X.shape[0]) should be wrapped in an array format,
like np.ones((X.shape[0], 1)),
to create a column vector."""

#x_with_intercept=np.concatenate(np.ones(X.shape[0]),X)


x_with_intercept=np.concatenate([np.ones((X.shape[0],1)),X],axis=1)
print(x_with_intercept)
```

```python
#or you can use this 
X_with_intercept = np.c_[np.ones(X.shape[0]), X]
X_with_intercept
```

```python
x_matrix=np.array(x_with_intercept)
y_matrix=np.array(Y)
```

```python
print(x_matrix)
print(y_matrix)
```

```python
theta = np.linalg.inv(x_matrix.T @ x_matrix) @ x_matrix.T @ y_matrix
```

```python
intercept = theta[0]
slopes = theta[1:]
```

```python
print("Intercept (c):", intercept)
print("Slopes (m1, m2, ..., mn):", slopes)
```

```python
Y_pred = x_matrix @ theta
print(Y_pred)
```

```python
m = len(y_matrix)  # Number of data points
J_theta = (1 / (2 * m)) * np.sum((Y_pred - y_matrix) ** 2)
```

```python
J_theta
```

```python
# Initialize parameters (theta) to zeros or small random values
theta = np.zeros(x_matrix.shape[1])

# Set learning rate and number of iterations
alpha = 0.01  # learning rate
iterations = 1000  # number of iterations

# Gradient Descent Algorithm
for _ in range(iterations):
    # Compute predicted values
    Y_pred = x_matrix @ theta
    
    # Compute the gradient
    gradient = (1 / len(y_matrix)) * x_matrix.T @ (Y_pred - y_matrix)
    
    # Update the parameters (theta)
    theta -= alpha * gradient

# Print the final theta values (intercept and slopes)
print("Trained parameters (theta):", theta)
```
