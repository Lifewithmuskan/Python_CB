```python
def hypothesis(X,weights):
    y_pred= weights[0]+weights[1]*X[0]
    return y_pred
```

```python
def loss_func(X,weight,Y):
    error=0
    for i in range(X.shape[0]):
        y_pred=hypothesis(X[i],weight)
        error+=(y_pred-Y[i])**2
    return error/X.shape[0]
```

```python
def gradient(X,weights,Y):
    grad=np.zeros((2,))
    for i in range(X.shape[0]):
        y_pred=hypothesis(X[i],weights)
        
        grad[0]+=(y_pred -Y[i])
        grad[1]+=(y_pred-Y[i])*X[i]
    return grad/X.shape[0]
```
